```python
# Quiz APP using OOP.

# Question
class Question:
  def __init__(self,text,choices,answer):
    self.text = text
    self.choices = choices
    self.answer = answer

  def checkAnswer(self,answer):
    return self.answer == answer

# Quiz
class Quiz:
  def __init__(self,questions):
    self.questions = questions
    self.score = 0
    self.questionIndex = 0

  def getQuestion(self): 
    return self.questions[self.questionIndex]
  
  def displayQuestion(self): 
    question = self.getQuestion()
    print(f"Question {self.questionIndex + 1}: {question.text}")

    for i in question.choices:
      print("-"+ i)

    answer = input("Answer: ")
    self.guess(answer)
    self.loadQuestion()
  
  def guess(self,answer):
    question = self.getQuestion()

    if question.checkAnswer(answer):
      self.score +=1
    self.questionIndex+=1
  
  def loadQuestion(self):
    if len(self.questions) == self.questionIndex:
      self.showScore()
    else:
      self.displayProgress()
      self.displayQuestion()

  def showScore(self):
    print("Score: ",self.score)

  def displayProgress(self):
    totalQuestion = len(self.questions)
    questionNumber = self.questionIndex + 1

    if questionNumber > totalQuestion:
      print("Quiz is over.")
    else:
      print(f"Question {questionNumber} of {totalQuestion}".center(100,"*"))


# Amount of questions can be added more according to quizmaker
q1 = Question("Question text ?",["Option a","Option b","Option c","Option d","Option e"],"Correct answer :a")

q2 = Question("Question text ?",["Option a","Option b","Option c","Option d","Option e"],"Correct answer :b")

q3 = Question("Question text ?",["Option a","Option b","Option c","Option d","Option e"],"Correct answer :c")

q4 = Question("Question text ?",["Option a","Option b","Option c","Option d","Option e"],"Correct answer :d")

q5 = Question("Question text ?",["Option a","Option b","Option c","Option d","Option e"],"Correct answer :e")


questions = [q1,q2,q3,q4,q5]

quiz = Quiz(questions)

quiz.loadQuestion()
```
